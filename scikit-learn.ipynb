## 监督学习
### 线性模型
`sklearn.linear_model`模块实现了线性模型.  
训练数据集$T=\{(x_1,y_1),(x_2,y_2),...,(x_N,y_N)\}$，其中$x_i=(x_{i1},x_{i2},...,x_{in})^T$为n维输入特征向量，$y_i$为标量输出。  
线性回归模型的函数表示：
$$\widehat{y_i}(\omega,x_i)=\omega_0+\omega_1x_{i1}+...+\omega_nx_{in}=\vec{\omega}^T·\vec{\overline x_i}$$
其中，$\vec{\overline x_i}=(1,x_{i1},x_{i2},...,x_{in})$，$\vec{\omega}=(\omega_0,\omega_1,\omega_2,...,\omega_n)$  
线性回归模型的损失函数定义为：
$$L=\frac{1}{N}\sum_{i=1}^N(y_i-\widehat{y})^2=\frac{1}{N}\sum_{i=1}^N[y_i-\widehat{y_i}(\omega,x_i)]^2$$

1. 普通最小二乘法  
   `LinearRegression`类实现了普通最小二乘线性回归模型，普通最小二乘法就是通过学习，求解使得损失函数最小的参数$\omega$，即$\mathop{min}\limits_{\omega} L$   
   最小二乘法主要用于求解多元函数的极值问题，原理是对损失函数L对参数$\omega$求偏导，并使偏导为0，然后解方程组.  
   线性模型对应的数学方程组可写成$X\vec{\omega}=\vec{y}$，即
   $$\begin{bmatrix}
      1&x_{11}&x_{12}&...&x_{1n}\\
      1&x_{21}&x_{22}&...&x_{2n}\\
      ...&...&...&...&...\\
      1&x_{N1}&x_{N2}&...&x_{Nn}\\
      \end{bmatrix}
      \begin{bmatrix}
      \omega_0\\
      \omega_1\\
      ...\\
      \omega_n\\
      \end{bmatrix}=
      \begin{bmatrix}
      y_1\\
      y_2\\
      ...\\
      y_N\\
      \end{bmatrix}$$
   如果输入向量的各特征之间线程无关，则可求得参数结果为：
   $$\omega=(X^TX)^{-1}X^Ty$$
   如果输入向量的各特征之间线性相关，则可以用奇异值分解（singular value decomposition）求得参数结果为：
   $$\omega=Q(\sum)^{-1}P^Ty$$
   > **对称矩阵对角化**  
   > 方阵$A$是对称矩阵，即满足$A^T=A$，则可分解为：$A=Q\Lambda Q^T$，$Q$为正交阵（$Q^{-1}=Q^T$）  
   > **奇异值分解**   
   > 矩阵$M_{m\times n}(m\neq n)$不是方阵，但矩阵$(AA^T)_{m\times m}$和$(A^TA)_{n\times n}$是对称方阵，可以进行分解：  
   > $$AA^T=P\Lambda_1P^T,A^TA=Q\Lambda_2Q^T$$  
   > 则矩阵$M$的奇异值分解为$M=P\sum Q^T$，其中，$P_{m\times m}$是酉矩阵；$\sum_{m\times n}$是非负实数对角矩阵（对角线元素为M的奇异值）；$Q_{n\times n}^T$是酉矩阵.  
   > 此时，矩阵$M$的伪逆矩阵为$M^{-1}=(P\sum Q^T)^{-1}=Q(\sum)^{-1}P^T$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets as skd
import sklearn.linear_model as skl

# 加载数据集并选择数据集的一个特征
data_X,data_y=skd.load_diabetes(return_X_y=True)
data_X=data_X[:,np.newaxis,2]

# 定义训练集和测试集
data_trainX=data_X[:-20]
data_trainy=data_y[:-20]
data_testX=data_X[-20:]
data_testy=data_y[-20:]

# 创建线性模型并进行学习
exp_model=skl.LinearRegression()
exp_model.fit(data_trainX,data_trainy)

# 利用测试集进行测试
predict_y=exp_model.predict(data_testX)

# 绘图
plt.scatter(data_testX,data_testy,label='test data')
plt.plot(data_testX,predict_y,'r-',label='predict data')
plt.legend()

print(exp_model.coef_)
print(exp_model.intercept_)
```

2. 岭（Ridge）回归和分类  
   `Ridge`类实现了岭回归，当数据特征中存在共线性时，原始最小二乘法会出现较大误差，岭回归在原始最小二乘法基础加入系数$w$二阶范数的正则化项（L2正则化），即：
   $$\mathop{min}\limits_\omega\{\frac{1}{N}||X\omega-y||_2^2+\alpha||\omega||_2^2\}$$

   `RidgeClassifier`类实现了岭回归分类，可以完成二分类和多分类任务，在二分类中分类器将数据标注转化为二元目标{-1,1}，然后用回归模型实现二分类，即根据回归器输出值的符号进行分类；在多分类中，将回归的输出值进行区间划分，并根据预测值所在区间进行分类.

   `GidgeCV`类实现了对$\alpha$的交叉验证，默认为留一交叉验证，可以通过指定属性值$cv$改变验证方式.

```python
# 创建希尔伯特矩阵（高度病态矩阵，任一元素发生变化对矩阵产生巨大影响）
X=1/(np.arange(1,11)+np.arange(0,10)[:,np.newaxis])
y=np.ones(10)

n_alphs=200
alphs=np.logspace(-10,-2,n_alphs)

# 对不同的正则化参数进行拟合并得到参数
coefs=list()
for i in alphs:
    ridge=skl.Ridge(alpha=i,fit_intercept=False)
    ridge.fit(X,y)
    coefs.append(ridge.coef_)

# 从结果可以看出来，α接近于0时，系数将非常震荡
ax=plt.gca()

ax.plot(alphs,coefs)
ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1])
```

3. Lasso  
   `Lasso`类实现了Lasso线性模型，在原始线性模型中加入参数$\omega$一阶范数的正则化项（L1正则化），即：
   $$\mathop{min}\limits_\omega\{\frac{1}{N}||X\omega-y||_2^2+\alpha||\omega||_1\}$$
   Lasso主要用于估计稀疏系数，使用**坐标下降法**拟合参数，可以用于**特征选择**.

   正则化参数$\alpha$可以使用交叉验证和基于信息准则的模型选择进行确定：  
   - `LassoCV`和`LassoLarsCV`类实现了对参数$\alpha$的交叉验证，前者适用于具有共线性特征的高维数据集，后者基于最小角回归算法实现交叉验证，在样本数量较小时使用.
   - `LassoLarsIC`类提出了2种基于信息准则的模型选择：赤池信息准则Akaike information criterion (AIC)和贝叶斯信息准则the Bayes Information criterion (BIC)，该方式假设模型必须是正确的，不适用于背景条件很差（特征数大于样本数）的问题.

   > **L1正则化和L2正则化**  
   > **坐标下降法(coordinate descent)**  
   > 坐标下降法是依次沿着坐标轴的方向最小化目标函数值，即将高维优化问题分解为多个一维的优化问题。  
   > 当多个变量之间存在较强的关联性时，应首先使用主成分分析法（principle components analysis,PCA）使各个变量之间尽可能相互独立。

```python
import time
# 加载数据集
X,y=skd.load_diabetes(return_X_y=True)

# 随机生成一组特征数据（bad features）并对所有特征归一化处理
rng=np.random.default_rng(seed=42)
X=np.c_[X,rng.standard_normal((X.shape[0],14))]
X/=np.sqrt(np.sum(X**2,axis=0))

# 基于信息准则的模型选择方法确定正则化参数
# AIC
model_aic=skl.LassoLarsIC(criterion='aic',normalize=False)
t1=time.time()
model_aic.fit(X,y)
aic_alpha=model_aic.alpha_
aic_time=time.time()-t1
# BIC
model_bic=skl.LassoLarsIC(criterion='bic',normalize=False)
model_bic.fit(X,y)
bic_alpha=model_bic.alpha_

# AIC和BIC的图像
plt.figure()
plt.semilogx(model_aic.alphas_,model_aic.criterion_,'r--',label='AIC')
plt.axvline(model_aic.alpha_,color='r',label='AIC alpha')

plt.semilogx(model_bic.alphas_,model_bic.criterion_,'b--',label='BIC')
plt.axvline(model_bic.alpha_,color='b',label='BIC alpha')

plt.xlabel(r'$\alpha$')
plt.ylabel('criterion')
plt.legend()
plt.title('Information-criterion for model selection(training time %.3fs)' % aic_time)

# 使用LassoCV(坐标下降法)确定正则化参数
model_cv=skl.LassoCV(cv=20)
t2=time.time()
model_cv.fit(X,y)
cv_time=time.time()-t2

plt.figure()
plt.semilogx(model_cv.alphas_,model_cv.mse_path_,':')
plt.plot(model_cv.alphas_,model_cv.mse_path_.mean(axis=1),'b',label='Average across the folds.')
plt.axvline(model_cv.alpha_,color='k',label='CV alpha')
plt.axis('tight')
plt.ylim(2300,3800)

plt.xlabel(r'$\alpha$')
plt.ylabel('Mean square error')
plt.legend()
plt.title('Coordinate descent(training time %.3fs)' % cv_time)

# 使用LassoLarsCV(最小角回归法)确定正则化参数
model_lars=skl.LassoLarsCV(cv=20,normalize=False)
t3=time.time()
model_lars.fit(X,y)
lars_time=time.time()-t3

plt.figure()
plt.semilogx(model_lars.cv_alphas_,model_lars.mse_path_,':')
plt.semilogx(model_lars.cv_alphas_,model_lars.mse_path_.mean(axis=-1),'b',label='Average across the folds.')
plt.axvline(model_lars.alpha_,color='k',label='CV alpha')
plt.axis('tight')
plt.ylim(2300,3800)

plt.xlabel(r'$\alpha$')
plt.ylabel('Mean square error')
plt.legend()
plt.title('Least angle regression(training time %.3fs)' % lars_time)
```

4. 多任务Lasso  
   `MultiTaskLasso`类实现了多任务联合回归问题的稀疏系数估计($y_{n_{samples}\times n_{tasks}}$)，主要在线性模型中加入了混合正则化项($L_1L_2范数$)，利用坐标下降法拟合参数.
   $$\mathop{min}\limits_\omega\{\frac{1}{2n_{samples}}||XW-Y||_{Fro}^2+\alpha||W||_{21}\}$$
   其中，Fro表示矩阵范数，即：
   $$||A||_{Fro}=\sqrt{\sum_{ij}a_{ij}^2}$$
   $L_1L_2$范数：
   $$||A||_{21}=\sum_i\sqrt{\sum_ja_{ij}^2}$$

5. 弹性网络(Elastic-Net)  
   `ElasticNet`类实现了弹性网络线性模型，是对Ridge和Lasso模型的结合，主要用于多个输入特征彼此相互关联的情况：
   $$\mathop{min}\limits_\omega\{\frac{1}{2n_{samples}}||X\omega-y||_2^2+\alpha\rho||W||_1+\frac{\alpha(1-\rho)}{2}||\omega||_2^2\}$$
   `ElasticNetCV`类实现了交叉验证确定参数$\alpha$和$\rho$.

7. 最小角回归(Least Angle regression)  
   `Lars`类实现了最小角回归算法，该算法是对前项选择算法和前项梯度算法的折中.  
   首先，找到与$Y$最接近的$X_k$，使用类似前项梯度算法沿该方向移动找到另一$X_t$，此时残差$Y_{res}$与$X_k$和$X_t$的相似程度一致(即残差位于$X_k$和$X_t$的角平分线上)；然后沿着残差方向继续前进，直到下一个向量相似度一致的向量.
   

   **前项选择算法(Forward selection)**  
   对于线性关系：
   $$X_{m\times n}\omega_{n\times 1}=Y_{m\times 1}=\begin{bmatrix}
    X_1&X_2&...&X_n
   \end{bmatrix}\omega$$
   其中m为样本数量，n为特征维度，选择与$Y$最为接近的$X_k$(余弦相似度最大)逼近$Y$，即：
   $$\overline Y=Y\theta_k,\theta_k=\frac{<X_k,Y>}{||X_k||_2·||Y||_2}$$
   此时的残差定义为:$Y_{res}=Y-\overline Y$，使用$Y_{res}$作为新的$Y$，去掉$X_k$后重复上述操作，直至残差为0或者所有的$X_i$全部使用，停止算法.

   **前向梯度算法**   
   前向梯度算法选择与$Y$最为接近的$X_k$(余弦相似度最大)，沿着$X_k$的方向移动固定值，然后计算残差$Y_{res}$，接着把残差作为新的变量重复上述操作，直至残差足够小时停止算法.

### 数据预处理(preprocessing data)
`sklearn.preprocessing`模块提供了数据集预处理的函数和类，数据集预处理主要包括7个关键步骤：获取数据集；导入数据处理包；导入数据集；缺失值处理；编码分类数据；划分数据集；特征缩放.
1. 标准化(Standardization)、均值消除(mean removal)、方差缩放(variance scaling)
   `StandardScaler`类通过公式$z=\frac{x-u}{s}$实现数据标准化，其中，$u$为样本平均值，$s$为样本标准差，标准化后的数据均值为0，方差为1.
