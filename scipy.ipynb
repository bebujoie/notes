#### 1.3.3 maskedarray: dealing with (propagation of) missing data
A masked array is the combination of a standard numpy.ndarray and a mask. A mask is either nomask, indicating that no value of the associated array is invalid, or an array of booleans that determines for each element of the associated array whether the value is valid or not. When an element of the mask is False, the corresponding element of the associated array is valid and is said to be unmasked. When an element of the mask is True, the corresponding element of the associated array is said to be masked (invalid).

#### 2.2.9 Moving spines
Spines are the lines connecting the axis tick marks and noting the boundaries of the data area. They can be placed at arbitrary positions.
#### 2.2.10 Devil is in the details
We can make the tick labels bigger and we can also adjust their properties such that they’ll be rendered on a semi-transparent white background. This will allow us to see both the data and the labels.

```python
# Create a figure of size 8x6 inches, 80 dots per inch
plt.figure(figsize=(8,6),dpi=80)
# Create a new subplot from a grid of 1x1
plt.subplot(1,1,1)
x=np.linspace(-2*np.pi,2*np.pi,256)
c,s=np.cos(x),np.sin(x)
# Plot with a blue continuous line of width 1 (pixels)
plt.plot(x,c,color='blue',linewidth=1.0,linestyle='-',label='cos')
# Plot with a green continuous line of width 1 (pixels)
plt.plot(x,s,color='green',linewidth=1.0,linestyle='-',label='sin')
# Set limits
plt.xlim(-2*np.pi,2*np.pi)
plt.ylim(-1.0,1.0)
# Set ticks and Set tick labels
plt.xticks(np.linspace(-2*np.pi,2*np.pi,5),[r'$-2\pi$',r'$-\pi$',r'$0$',r'$\pi$',r'$2\pi$'])
plt.yticks(np.linspace(-1.0,1.0,5))

# Moving spines
ax = plt.gca() # gca stands for 'get current axis'
ax.spines['left'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('right')
ax.spines['right'].set_position(('data',0))

# Annotate some points
t=2*np.pi/3
plt.plot([t,t],[np.sin(t),np.cos(t)],'r--o')
plt.annotate(r'$y=sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',xy=(t,np.sin(t)),
            xycoords='data',textcoords='offset points',
            xytext=(10,30),arrowprops=dict(arrowstyle='->'))
# Create a lengend
plt.legend()
# 
# Save figure using 72 dots per inch
# plt.savefig("exercise_2.png", dpi=72)
```

### 2.4 Other Types of Plots
#### 2.4.1 Regular Plots

```python
n=256
X=np.linspace(-np.pi,np.pi,n)
Y=np.sin(2*X)

plt.axes([0.025,0.025,0.95,0.95])
plt.plot(X,Y+1,'b-',alpha=1.0)
plt.fill_between(X,1,Y+1,color='blue',alpha=0.25)

plt.plot(X,Y-1,'b-',alpha=1.0)
# Fill the area between two horizontal curves.
plt.fill_between(X,-1,Y-1,(Y-1)<-1,color='blue',alpha=0.25)
plt.fill_between(X,-1,Y-1,(Y-1)>-1,color='red',alpha=0.25)
```

#### 2.4.2 Scatter Plots

```python
# A scatter plot of y vs. x with varying marker size and/or color.
# scatter(x, y, s=None, c=None, marker=None, cmap=None, 
#       norm=None, vmin=None, vmax=None, alpha=None, 
#       linewidths=None, *, edgecolors=None, plotnonfinite=False, 
#       data=None, **kwargs)
# s:The marker size in points**2. Default is rcParams['lines.markersize'] ** 2.
# c:The marker colors.
# alpha:The alpha blending value, between 0 (transparent) and 1 (opaque).
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

points=30*(X**2);
colors=np.random.rand(n)
plt.scatter(X,Y,s=points,c=colors,alpha=0.5)
```
